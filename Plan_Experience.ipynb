## 1. Pourquoi un plan d‚Äôexp√©rience ?

> Un **plan d‚Äôexp√©rience** est une m√©thode qui consiste √† :

## 3. Notions cl√©s √† conna√Ætre

### Facteur

## 4. Limite de la m√©thode ‚Äúun param√®tre √† la fois‚Äù

Faire varier un param√®tre √† la fois peut sembler logique,

> Une interaction signifie que l‚Äôeffet d‚Äôun param√®tre d√©pend

## 5. Principe du plan factoriel simple (2 facteurs)

Avec deux facteurs A et B √† deux niveaux (bas / haut),

| Essai | Facteur A | Facteur B |

### Mais attention : limite fondamentale (aliasing)

> ‚ÄúAvec seulement 4 essais, nous proposons un plan fractionnaire structur√© permettant
> de comparer des configurations contrast√©es et d‚Äôobtenir une premi√®re indication des param√®tres influents.
> Ce plan est un compromis co√ªt/information : il peut r√©v√©ler des tendances fortes,
> mais certaines conclusions restent ambigu√´s car des interactions sont confondues avec les effets principaux.
> Si un facteur semble dominant, une seconde phase d‚Äôessais sera n√©cessaire pour confirmer

  - soit on planifie une phase 2 (plan complet ou essais cibl√©s)

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# ------------------------------------------------------------
# Plan fractionnaire 2^(3-1) : 4 essais
# Codage : -1 = niveau bas, +1 = niveau haut
# Essais propos√©s :
# 1: A- B- C-
# 2: A+ B- C+
# 3: A- B+ C+
# 4: A+ B+ C-
# ------------------------------------------------------------
design = pd.DataFrame({
    "A_vitesse": [-1, +1, -1, +1],
    "B_avance":  [-1, -1, +1, +1],
    "C_lubr":    [-1, +1, +1, -1],
})

# Pour affichage "lisible" (optionnel)
design_readable = design.replace({-1: "Bas", +1: "Haut"})
design_readable.index = [f"Essai {i}" for i in range(1, 5)]
display(design_readable)

# ------------------------------------------------------------
# Simulation d'une r√©ponse (usure en mm) pour l'exemple
# IMPORTANT : c'est une simulation p√©dagogique, pas une v√©rit√© industrielle
# y = base + effetA*A + effetB*B + effetC*C + bruit
# ------------------------------------------------------------
rng = np.random.default_rng(0)
base = 0.30
effA = 0.05   # ex : haute vitesse -> +usure
effB = 0.03   # ex : forte avance  -> +usure
effC = -0.04  # ex : lubrification pr√©sente -> -usure
sigma = 0.005 # petit bruit (mais avec 4 essais, √ßa ne "sauve" pas le ddl)

y = (base
     + effA*design["A_vitesse"]
     + effB*design["B_avance"]
     + effC*design["C_lubr"]
     + rng.normal(0, sigma, size=len(design)))

design["Usure_mm"] = y
display(design)

# ------------------------------------------------------------
# Ajustement r√©gression sur effets principaux (A, B, C)
# ATTENTION : avec 4 essais et (intercept + 3 effets) = 4 param√®tres
# -> mod√®le satur√© : ajustement parfait, ddl r√©siduel = 0
# -> p-values et tests classiques ne sont pas interpr√©tables
# ------------------------------------------------------------
X = sm.add_constant(design[["A_vitesse", "B_avance", "C_lubr"]])
model = sm.OLS(design["Usure_mm"], X).fit
print(model.summary())

# Affiche uniquement les coefficients (OK m√™me si mod√®le satur√©)
print("Coefficients estim√©s (mod√®le satur√©, pas de p-values exploitables) :")
print(model.params)

print("\n‚ö†Ô∏è Note : avec 4 essais et 4 param√®tres (constante + A + B + C),")
print("ddl r√©siduels = 0 ‚Üí impossible d'estimer l'erreur ‚Üí pas de tests/p-values fiables.")


print("\n‚ö†Ô∏è Remarque p√©dagogique :")
print("- Ici, 4 essais = mod√®le satur√© (ddl r√©siduel = 0).")
print("- On peut estimer des effets, mais on ne peut pas estimer correctement l'erreur.")
print("- En pratique, on ajoute des r√©p√©titions (r√©plicats) ou des points centre pour faire de l'inf√©rence.")
```

## Interpr√©tation d√©taill√©e des r√©sultats du plan d‚Äôexp√©rience

Les coefficients ci-dessous proviennent de l‚Äôajustement d‚Äôun mod√®le lin√©aire
sur un **plan d‚Äôexp√©rience fractionnaire** √† trois facteurs, cod√©s selon la convention

Le mod√®le utilis√© est :

o√π $Y$ repr√©sente l‚Äôusure de l‚Äôoutil (en mm).

---

### 1. Interpr√©tation de la constante (intercept)

**Constante :**
$$
\beta_0 = 0.300924

Cette valeur correspond √† l‚Äôusure **pr√©dite au centre du domaine exp√©rimental**,

Pour obtenir l‚Äôeffet r√©el entre le niveau bas et le niveau haut,

**Interpr√©tation ing√©nieur :**

#### Effet de l‚Äôavance (facteur B)

**Interpr√©tation ing√©nieur :**

|-------|-------------------------|---------------------|

### 4. Limite d‚Äôinterpr√©tation (plan fractionnaire)

Ce plan d‚Äôexp√©rience comporte seulement 4 essais et ne permet pas de s√©parer

En particulier :
- l‚Äôeffet de la vitesse est confondu avec l‚Äôinteraction avance √ó lubrification,
- l‚Äôeffet de l‚Äôavance est confondu avec l‚Äôinteraction vitesse √ó lubrification,

### 5. Conclusion d‚Äôing√©nieur

Le plan fractionnaire met en √©vidence des effets importants associ√©s
